```python
# Reading in excel spreadsheet

import pandas as pd

data = '2023.11 Invoice Distributing.xlsx'

df = pd.read_excel(data)
df
```

```python
# creating PI + account identifier, where applicable

pi_account = []
for i, row in df.iterrows():
    if df['Account'][i] != 'unknown':
        pi_account.append(f"{df['PI'][i]}_{df['Account'][i]}")
    elif df['Account'][i] == 'unknown':
        pi_account.append(f"{df['PI'][i]}")
df['PI_Account'] = pi_account
df
```

```python
# creating new dataframe to total uses and charge

df2 = pd.DataFrame()
df2['PI_Account'] = df['PI_Account'].unique()
df2['Number Uses'] = 0
df2['Total ($)'] = 0

df2
```

```python


for i, row in df2.iterrows():
    owed = []
    num_uses = []
    
    for j, rows in df.iterrows():
        if df2['PI_Account'][i] == df['PI_Account'][j]:
            owed.append(df['Charge'][j])
            num_uses.append(df['Fraction'][j])
        # print(owed)

        df2['Number Uses'][i] = sum(num_uses)
        df2['Total ($)'][i] = sum(owed)

df2 = df2.round(2)
    
```

```python
df2.to_excel('ethylene_oxide_invoice_total_through_November_2023.xlsx', index=False)
```
